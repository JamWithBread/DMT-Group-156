```python
import seaborn as sns
import matplotlib.pyplot as plt
# ID of country hotel is located in
plt.figure(figsize = (22,8))
plt.xticks(rotation = -60, size =7.5, weight = 'bold')
sns.countplot(x='prop_country_id', data=df_train)
```

```python
#Count by country of visitor origin
plt.figure(figsize = (25,8))
plt.xticks(rotation = -60, size =7.5, weight = 'bold')
sns.countplot(x='visitor_location_country_id', data=df_train)
```

```python
nulls = (df_train.isnull().sum()/len(df_train))*100
#nulls
```

```python
nulls = pd.DataFrame(nulls)
```

```python
nulls.reset_index(inplace=True)
```

```python
nulls = nulls.sort_values(by = 0, ascending = False)
nulls
```

```python
# Percentage of missing values per feature
fig = plt.figure(figsize=(20,5))

ax = fig.add_axes([0,0,1,1])
features = nulls['index']
values = nulls[0]
ax.bar(features,values)
plt.xticks(rotation = 90)
plt.title("Missing Values by Feature")
plt.ylabel("Percent Missing")
plt.xlabel("Feature")

plt.show()
```

```python
#srch_children_count: The number of children specified in the hotel room
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='srch_children_count', data=df_train)
```

```python
#srch_adults_count: The number of adults specified in the hotel room
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='srch_adults_count', data=df_train)
```

```python
#prop_review_score: The mean customer review score for the hotel on a scale out of 5, rounded to 0.5 increments. 
# A 0 means there have been no reviews, null that the information is not available
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='prop_review_score', data=df_train)
```

```python
# prop_brand_bool: +1 if the hotel is part of a major hotel chain; 0 if it is an independent hotel
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='prop_brand_bool', data=df_train)
```

```python
#srch_room_count: Number of hotel rooms specified in the search
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='srch_room_count', data=df_train)
```

## Expedia vs competitors on price, availability and % diff vs competitor's price
#### These 3 visualizations could be performed for all 8 competitors

```python
#comp1_rate: +1 if Expedia has a lower price than competitor 1 for the hotel; 0 if the same;
# -1 if Expediaâ€™s price is higher than competitor 1; null signifies there is no competitive data
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='comp1_rate', data=df_train)
```

```python
#comp1_inv: +1 if competitor 1 does not have availability in the hotel; 
# 0 if both Expedia and competitor 1 have availability; null signifies there is no competitive data
plt.figure(figsize = (10,5))
plt.xticks(rotation = 0, size =20)
sns.countplot(x='comp1_inv', data=df_train)
```

```python
#comp1_rate_percent_diff: The absolute percentage difference (if one exists) between Expedia and 
# competitor 1â€™s price (Expediaâ€™s price the denominator); null signifies there is no competitive data
plt.figure(figsize = (28,5))
plt.xticks(rotation = -60, size =7.5)
sns.countplot(x='comp1_rate_percent_diff', data=df_train)
```
