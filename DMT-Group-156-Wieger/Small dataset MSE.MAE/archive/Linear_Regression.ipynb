```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('Fish.csv',",")
#df_plots = df.copy()
```

```python
df
```

```python
groups = df.groupby("Species")
```

```python
fig1 = plt.figure()

group_counts = groups["Species"].count()

fig1 = plt.bar(group_counts.keys(), list(group_counts))

fig1.
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot()

fig2 = plt.figure()
ax2 = fig2.add_subplot()

fig3 = plt.figure()
ax3 = fig3.add_subplot()

for name, group in groups:
    ax1.plot(group.Length1, group.Height, marker="o", linestyle="", label = name)
    ax2.plot(group.Length1, group.Width, marker="o", linestyle="", label = name)
    ax3.plot(group.Length1, group.Weight, marker="o", linestyle="", label = name)
    
ax1.set_ylabel("Height (cm)")
ax1.set_xlabel("Length1 (cm)")

ax2.set_ylabel("Width (cm)")
ax2.set_xlabel("Length1 (cm)")

ax3.set_ylabel("Weight (g)")
ax3.set_xlabel("Length1 (cm)")

fig1.legend()
fig2.legend()
fig3.legend()
```

```python
fig = []
for name, group in groups:
    fig.append(plt.figure())
    ax = fig[-1].add_subplot()
    ax.plot(group.Length1, group.Weight, marker="o", linestyle="")
    fig[-1].suptitle(name)
```
